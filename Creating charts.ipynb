```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

with open('radishsurvey.txt') as readfile:
    name_radish = [line.split('-') for line in readfile.readlines()]
    radish_names = [entry[1] for entry in name_radish]
    radish_names_cleaned = [name.strip().replace('\s+',' ').lower() for name in radish_names]
    counts = Counter()
    for name in radish_names_cleaned:
        counts[name] += 1
    print(counts)
```

```python
    names = counts.keys()
    votes = counts.values()
    
    x = np.arange(len(counts))
    
    %matplotlib inline
    plt.bar(x, votes)
    plt.xticks(x + 0.5, names, rotation=90)
    
    print("hello world!")
```
