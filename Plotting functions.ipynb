```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-2, 2, 100, endpoint = True)

def add_splines(ax):
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
```

```python
# Linear
def linear(xval):
    return 5*xval - 3
Y = [linear(xval) for xval in X]
plt.plot(X, Y, label='Linear')
plt.legend(loc='upper right')

ax = plt.gca()
add_splines(ax)
```

```python
# Quadratic
def quadratic(xval):
    return 3*xval*xval - 2*xval + 5

Y = [quadratic(xval) for xval in X]
plt.plot(X, Y, label='Quadratic')
plt.legend(loc='upper right')

ax = plt.gca()
add_splines(ax)
```

```python
# power
def power(xval, expval):
    return 2*(pow(xval, expval))

plt.figure(1)
plt.subplot(211)
Y = [power(xval, 3) for xval in X]
plt.plot(X, Y, label='Power 3')
plt.legend(loc='upper right')
ax = plt.gca()
add_splines(ax)

plt.subplot(212)
Y = [power(xval, -1) for xval in X]
plt.plot(X, Y, label='Power -1')
plt.legend(loc='upper right')

ax = plt.gca()
add_splines(ax)
```

```python
# Polynomial
def polynomial(xval):
    return 4*pow(xval,4) + 3*pow(xval,3) - 2*pow(xval,2) + 9*xval - 16

Y = [polynomial(xval) for xval in X]
plt.plot(X, Y, label='Polynomial')
plt.legend(loc='upper right')

ax = plt.gca()
add_splines(ax)
```

```python
# Rational
def rational(xval):
    return (5*xval*xval + 2*xval + 7)/(xval*xval - 3*xval -14)
Y = [rational(xval) for xval in X]
plt.plot(X, Y, label='Rational')
plt.legend(loc='upper right')

ax = plt.gca()
add_splines(ax)
```

```python
# Exponential
def exponential(xval):
    return 0.03 * pow(5,xval)

Y = [exponential(xval) for xval in X]
plt.plot(X, Y, label='Exponential')
plt.legend(loc='upper right')

ax = plt.gca()
add_splines(ax)
```

```python
# Logarithmic
import math
def logarithmic(xval):
    if xval == 0:
        xval = 0.01
    return 1.5*math.log(xval + 2.001) + 10

Y = [logarithmic(xval) for xval in X]
plt.plot(X + 2.001, Y, label='Logarithmic')
plt.legend(loc='upper right')

ax = plt.gca()
add_splines(ax)
```

```python
# Sinusoidal
def sinusoidal(xval):
    return 2.5 * math.sin(xval*1.5 - 3)

Y = [sinusoidal(xval * 4) for xval in X]
plt.plot(X * 4, Y, label='Sinusoidal')
plt.legend(loc='upper right')

ax = plt.gca()
add_splines(ax)
```

```python
# Normal distribution
def normal(xval): #mu=0,sigma=0.5
    expval = -xval*xval/(2*0.5*0.5)
    return math.exp(expval)/(0.5*math.sqrt(2*math.pi))

Y = [normal(xval) for xval in X]
plt.plot(X, Y, label='Normal')
plt.legend(loc='upper right')

ax = plt.gca()
add_splines(ax)
```

```python
# logistic regression graphs
import math
def logistic(xval):
    return 1/(1 + math.exp(-1.0 * xval))

Y = [logistic(xval*2) for xval in X]
plt.plot(X*2, Y, label='Logistic')
plt.legend(loc='upper right')

ax = plt.gca()
add_splines(ax)
```
